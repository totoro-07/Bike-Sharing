## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana trend penggunaan bike sharing setiap tahun?

```python
hours_df['bln'] = hours_df['date'].dt.strftime('%b')
hours_df['bln'] = pd.Categorical(hours_df['bln'],
                               ordered=True,
                               categories=['Jan',
                                           'Feb',
                                           'Mar',
                                           'Apr',
                                           'May',
                                           'Jun',
                                           'Jul',
                                           'Aug',
                                           'Sep',
                                           'Oct',
                                           'Nov',
                                           'Dec'])

bike = hours_df[['year', 'total_count', 'bln']]
bike2011 = bike.loc[bike['year'] == 0]
bike2011 = bike2011[['bln', 'total_count']].groupby('bln', observed=False).sum().reset_index()

bike2012 = bike.loc[bike['year'] == 1]
bike2012 = bike2012[['bln', 'total_count']].groupby('bln', observed=False).sum().reset_index()


plt.figure(figsize = (10,5))

plt.plot(
    bike2011['bln'],
    bike2011['total_count'],
    label='2011',
    marker=".",
    linewidth=2,
)

plt.plot(
    bike2012['bln'],
    bike2012['total_count'],
    label='2012',
    marker=".",
    linewidth=2,
)

plt.xlabel('Bulan')
plt.ylabel('Jumlah pengguna')
plt.title('Trend Penggunaan Bike Sharing')
plt.legend()
plt.show()
```

### Pertanyaan 2: Bagaimana pengaruh kondisi cuaca terhadap persentase jumlah pengguna bike sharing?

```python
def cuaca(x):
  if x == 1:
    return 'Clear'
  elif x == 2:
    return 'Cloudy'
  elif x == 3:
    return 'Light Snow'
  else:
    return 'Heavy Rain'

hours_df['cuaca'] = hours_df['weathersit'].apply(cuaca)

weather = hours_df[['cuaca', 'total_count']].groupby('cuaca').sum().sort_values(by='total_count', ascending=False).reset_index()

plt.figure(figsize = (10, 5))
plt.title("Persentase Pengguna Bike Sharing Berdasarkan Keadaan Cuaca")
colors = ['#3498db', '#2ecc71', '#f1c40f', '#e74c3c']

plt.pie(
    weather['total_count'],
    labels=weather['cuaca'],
    autopct= '%1.1f%%',
    colors = colors,
    wedgeprops={'edgecolor': 'white'}
)

plt.show()
```

### Pertanyaan 3:  Bagaimana pengaruh musim terhadap persentase jumlah pengguna *bike sharing*?

```python
def musim(x):
  if x == 1:
    return 'Spring'
  elif x == 2:
    return 'Summer'
  elif x == 3:
    return 'Fall'
  else:
    return 'Winter'

hours_df['musim'] = hours_df['season'].apply(musim)
season = hours_df[['musim', 'total_count']].groupby('musim').sum().sort_values(by='total_count', ascending=False).reset_index()

colors = ['#3498db', '#2ecc71', '#f1c40f', '#e74c3c']
plt.title("Persentase Pengguna Bike Sharing Berdasarkan Musim", fontsize=14)

plt.pie(
    season['total_count'],
    labels=season['musim'],
    autopct='%1.1f%%',
    colors=colors,
    wedgeprops={'edgecolor': 'white'}
)

plt.show()
```

### Pertanyaan 4: Pada jam berapa jumlah peminjaman sepeda paling tinggi?

```python
hours_df['musim'] = hours_df['season'].apply(musim)

user_hour = hours_df[['season', 'hour', 'total_count']]
user_hour = user_hour[['hour', 'season', 'total_count']].groupby('hour').sum().reset_index()

fig = plt.figure(figsize=(10, 5))

plt.figure(figsize=(10, 5))

plt.plot(
    user_hour['hour'],
    user_hour['total_count'],
    marker="o",
    linestyle="-",
    linewidth=2,
    color="#FF5733",
)

plt.ylabel('Jumlah Pengguna', fontsize=12)
plt.xlabel('Jam', fontsize=12)
plt.title('Tingkat Penggunaan Bike Sharing setiap jam', fontsize=14)

plt.xticks(ticks=range(0, 24), labels=range(0, 24))  # Menjaga keterbacaan sumbu x

plt.show()
```

**Insight:**
1. Trend Penggunaan Bike Sharing Setiap Tahun
  * Penggunaan bike sharing meningkat pada tahun 2012 dibandingkan dengan 2011.
  * Pola penggunaan menunjukkan tren musiman, dengan peningkatan dari awal tahun hingga puncaknya di musim panas (Juni-Agustus), lalu menurun menjelang akhir tahun.
  * Faktor cuaca dan musim kemungkinan besar mempengaruhi tingkat penggunaan ini.
2. Pengaruh Kondisi Cuaca terhadap Persentase Pengguna
  * Mayoritas pengguna bike sharing memilih untuk bersepeda dalam kondisi cuaca cerah (71%).
  * Cuaca mendung (24.2%) juga masih cukup banyak digunakan, tetapi jumlahnya jauh lebih rendah dibandingkan cuaca cerah.
  * Kondisi hujan lebat atau salju ringan memiliki dampak negatif terhadap jumlah pengguna, dengan persentase yang sangat kecil (<5%).
  * Ini menunjukkan bahwa kondisi cuaca yang lebih baik cenderung meningkatkan penggunaan bike sharing.
3. Pengaruh Musim terhadap Jumlah Pengguna
  * Penggunaan bike sharing paling tinggi terjadi pada musim gugur (32.2%) dan musim panas (27.9%).
  * Musim dingin memiliki tingkat penggunaan yang lebih rendah (25.6%), sedangkan musim semi memiliki penggunaan terendah (14.3%).
  * Ini bisa disebabkan oleh faktor cuaca yang lebih nyaman untuk bersepeda di musim gugur dan musim panas.
4. Jam dengan Jumlah Pemakaian Tertinggi
  * Pemakaian bike sharing meningkat secara signifikan pada jam sibuk, yaitu sekitar pukul 7-9 pagi dan 5-7 sore.
  * Peningkatan ini menunjukkan bahwa mayoritas pengguna kemungkinan besar adalah pekerja atau pelajar yang menggunakan bike sharing sebagai moda transportasi utama atau tambahan.
  * Selama tengah malam hingga dini hari (00:00 - 06:00), penggunaan sangat rendah.
  * Pola ini menunjukkan bahwa bike sharing lebih banyak digunakan untuk aktivitas sehari-hari seperti pergi dan pulang kerja/sekolah dibandingkan penggunaan rekreasional.
